```python
import pandas as pd
from matplotlib import pyplot as plt
import seaborn as sns
```

```python
ipl = pd.read_csv('matches.csv')
```

```python
# Having a glance of the first 5 items.
ipl.head()
```

```python
# To look at the number of rows and colums
ipl.shape
```

```python
# Looking for the most MOM awards
ipl['player_of_match'].value_counts()
```

```python
ipl['player_of_match'].value_counts().keys()
```

```python
# Getting the top 10 players with MOM awards
ipl['player_of_match'].value_counts()[0:10]
```

```python
# top 5
ipl['player_of_match'].value_counts()[0:5]
```

```python
# Listing the names of the players
list(ipl['player_of_match'].value_counts()[0:5].keys())
```

```python
# The Bar-Plot for the top 5 players with most man of the match
plt.figure(figsize=(8,5))
# 8,5 sets the dimensions of the image as l & b.
plt.bar(list(ipl['player_of_match'].value_counts()[0:5].keys()),list(ipl['player_of_match'].value_counts()[0:5]),color = "b")
# plt.bar plots the bar , it has 2 parameters frst id categorical value
# and second is numerical value
plt.show()
```

```python
plt.figure(figsize=(18,5))
plt.bar(list(ipl['player_of_match'].value_counts()[0:10].keys()),list(ipl['player_of_match'].value_counts()[0:10]),color = "r")
plt.show()
```

```python
# In the results column we are checking how many wins are by Runs, wickets and tie matches
ipl['result'].value_counts()
```

```python
# checking the teams that won the most number of toss
ipl['toss_winner'].value_counts()
```

```python
# checking the teams that won batting first
batting_first = ipl[ipl['result']== 'runs']
batting_first.head()
```

```python
batting_first
```

```python
# making a histogram since it is in numerical value
plt.figure(figsize=(7,7))
plt.hist(batting_first['result_margin'])
# Hist is used for making a histogram
plt.title("Distribution of runs")
plt.xlabel("Runs")
plt.show()
```

```python
# Finding out the number of wins w.r.t each team after batting first
batting_first['winner'].value_counts()
```

```python
# top 3 teams
batting_first['winner'].value_counts()[0:6].keys()
```

```python
list(batting_first['winner'].value_counts()[0:6])
```

```python
# making a bar-plot for top 6 teams with wins after batting first
plt.figure(figsize=(16,7))
plt.bar(list(batting_first['winner'].value_counts()[0:6].keys()),
        list(batting_first['winner'].value_counts()[0:6]),
        color=["blue","gold","indigo","tomato","red","tab:orange"])
plt.show()
```

```python
#making a pie chart
plt.figure(figsize=(7,7))
# on pie chart first is numerical value, then categorical entry
plt.pie(list(batting_first['winner'].value_counts()[0:11]),
        labels=list(batting_first['winner'].value_counts()[0:11].keys()),
         autopct='%0.1f%%')
# autopct is used to display the percentage value on the pie chart
# 0.1f is saying that we have .1 decimal values
plt.show()

```
